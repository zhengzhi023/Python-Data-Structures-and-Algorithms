```python
import numpy as np
import pandas as pd
import numpy
from datetime import datetime
file_path = open("./test.csv")
file_data = pd.read_csv(file_path)
file_data
```

```python
file_data = file_data.drop_duplicates()
file_data
```

```python
file_data = file_data.dropna()
file_data
```

```python
data_new = np.array([])
data = file_data['温度(℃)'].values
for i in data:
    data_new = np.append(data_new,np.array(i[:-1]))
data = data_new.astype(np.float64)
file_data.loc[:,'温度(℃)'] = data
file_data
```

```python
data_new = np.array([])
data = file_data['转速'].values
for i in data:
    data_new = np.append(data_new,np.array(i[:-1]))
data = data_new.astype(np.float64)
file_data.loc[:,'转速'] = data
file_data
```

```python
time_data = file_data['时间'].values
temp_list=[]
t10 = datetime.strptime('10:00','%H:%M')
t11 = datetime.strptime('11:00','%H:%M')
t12 = datetime.strptime('12:00','%H:%M')
t13 = datetime.strptime('13:00','%H:%M')
t14 = datetime.strptime('14:00','%H:%M')
t15 = datetime.strptime('15:00','%H:%M')
t16 = datetime.strptime('16:00','%H:%M')
t17 = datetime.strptime('17:00','%H:%M')
for j in time_data:
    i = datetime.strptime(str(j),'%H:%M')
    if (i<t10):
        temp_list.append("10点前")
    elif(t10<i and i<=t11):
        temp_list.append("10点至11点")
    elif(t11<i and i<=t12):
        temp_list.append("11点至12点")
    elif(t12<i and i<=t13):
        temp_list.append("12点至13点")
    elif(t13<i and i<=t14):
        temp_list.append("13点至14点")
    elif(t14<i and i<=t15):
        temp_list.append("14点至15点")
    elif(t15<i and i<=t16):
        temp_list.append("15点至16点")
    elif(t16<i and i<=t17):
        temp_list.append("16点至17点")
    else:
        temp_list.append("17点之后")
file_data.loc[:,'时间'] = temp_list
file_data
```

```python
df_all = pd.DataFrame({'时间':file_data['时间'].unique(),
                       '平均温度':0,
                       '平均转速':0})
df_all = df_all.sort_values(by=["时间"])
df_all
```

```python
import matplotlib.pyplot as plt
import matplotlib
av_wd = round(file_data['温度(℃)'].groupby(file_data['时间']).mean(),2)
av_zs = round(file_data['转速'].groupby(file_data['时间']).mean(),2)
df_all['平均温度']= av_wd.values
df_all['平均转速']= av_zs.values
df_all
```

```python
import matplotlib.ticker as mtick
from  matplotlib.font_manager import FontProperties
avwd = df_all['平均温度']
avzs = df_all['平均转速']
l = [i for i in range(9)]
plt.rcParams['font.sans-serif']=['SimHei']
lx = df_all['时间']
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(111)
ax1.plot(l,avwd,'or-',label='温度')

for i ,(_x,_y) in enumerate(zip(l,avwd)):
    plt.text(_x,_y,_y,color='black',fontsize=12)
ax1.set_ylim([0,100])
ax1.set_ylabel('温度')
plt.legend(prop={'family':'SimHei','size':10},loc='upper left')
ax2 = ax1.twinx()
plt.bar(l,avzs,alpha=0.5,color='green',label='转速')
ax2.set_ylim([0,8000])
ax2.set_label('转速')
plt.legend(prop={'family':'SimHei','size':10},loc='upper left')
plt.xticks(l,lx)
plt.show()
```

```python
data_new = []
data = file_data['日期'].values
for i in data:
    i = i.replace('年','-')
    i = i.replace('月','-')
    i = i.replace('日','')
    month = datetime.strptime(i, '%Y-%m-%d').month
    data_new.append(month)
file_data['日期'] = pd.Series(data_new).astype(np.int64)
file_data
```

```python
df_all=pd.DataFrame({'月份':file_data['日期'].unique(),
                     '平均温度':0,
                     '平均转速':0})
df_all = df_all.sort_values(by=["月份"])
df_all
```

```python
av_wd = round(file_data['温度(℃)'].groupby(file_data['日期']).mean(),2)
av_zs = round(file_data['转速'].groupby(file_data['日期']).mean(),2)
df_all['平均温度']= av_wd
df_all['平均转速']= av_zs
df_all
```

```python
avwd = df_all['平均温度']
avzs = df_all['平均转速']
# 获取月份的唯一值，并确保与l的长度相同
unique_months = df_all['月份'].unique()
l = list(range(len(unique_months)))

# 更新lx为与l相同长度的月份
lx = unique_months

# 继续绘图
fig = plt.figure(figsize=(10, 6))
ax1 = fig.add_subplot(111)
ax1.plot(l, avwd, 'or-', label='温度')
for i, (_x, _y) in enumerate(zip(l, avwd)):
    plt.text(_x, _y, _y, color='black', fontsize=12)
ax1.set_ylim([0, 100])
ax1.set_ylabel('温度')
plt.legend(prop={'family': 'SimHei', 'size': 10}, loc='upper left')

ax2 = ax1.twinx()
plt.bar(l, avzs, alpha=0.5, color='green', label='转速')
ax2.set_ylim([0, 8000])
ax2.set_ylabel('转速')  # 注意这里应该是set_ylabel而不是set_label
plt.legend(prop={'family': 'SimHei', 'size': 10}, loc='upper left')

plt.xticks(l, lx)
plt.show()
```

```python
print('电机的最高温度时%d度'%(max(file_data['温度(℃)'])))
print('电机的最低温度时%d度'%(min(file_data['温度(℃)'])))
```

```python
wd_divide = [1,35,40,50,60,70,80]
wd_cut = pd.cut(list(file_data['温度(℃)']),wd_divide)
wd_cut_data =  wd_cut.describe()
wd_cut_data
```

```python
area_percentage = (wd_cut_data['freqs'].values)*100
np.set_printoptions(precision=2)
lable = ['35度以下','35-40度','40-50度','50-60度','60-70度','70度以上',]
plt.axes(aspect=1)
plt.pie(x=area_percentage,labels=lable,autopct='%.2f %%',
        shadow=True,labeldistance=1.2,startangle=90,pctdistance=0.7)
plt.legend(loc='upper left',bbox_to_anchor=(1.5,1.2))
plt.show()
```

```python
print('电机的转速最高时%d'%(max(file_data['转速'])))
print('电机的转速最低时%d'%(min(file_data['转速'])))
```

```python
zs_divide = [1,3500,4000,5000,6000,7000,8000]
zs_cut = pd.cut(list(file_data['转速']),zs_divide)
zs_cut_data =  zs_cut.describe()
zs_cut_data
```

```python
area_percentage = (zs_cut_data['freqs'].values)*100
np.set_printoptions(precision=8)
lable = ['3500转以下','3500-4000转','4000-5000转','5000-6000rpm','6000-7000转','7000转以上',]
plt.axes(aspect=1)
plt.pie(x=area_percentage,labels=lable,autopct='%.2f %%',
        shadow=True,labeldistance=1.2,startangle=90,pctdistance=0.7)
plt.legend(loc='upper left',bbox_to_anchor=(1.5,1.2))
plt.show()
```
